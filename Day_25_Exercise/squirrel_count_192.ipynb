```python
import pandas as pd

# read in the data
df = pd.read_csv("2018_Central_Park_Squirrel_Census_-_Squirrel_Data_20251109.csv")
print(df.head())
```

```python
# look in the columns names
print(df.columns)
```

```python
print(df['Primary Fur Color'].value_counts())
```

```python
grey_squirrel_count = len(df.loc[df['Primary Fur Color']== "Gray"])
red_squirrel_count = len(df.loc[df['Primary Fur Color']== "Cinnamon"])
black_squirrel_count = len(df.loc[df['Primary Fur Color']== "Black"])

squirrel_count = [grey_squirrel_count, red_squirrel_count, black_squirrel_count]
fur_color = ['grey', 'red', 'black']

summed_df = pd.DataFrame({
    'Fur Color': fur_color,
    'Count': squirrel_count,
})

print(summed_df)
```

```python
df.to_csv("squirrel_count.csv")
```
